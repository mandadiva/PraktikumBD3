```python
from collections import defaultdict

dataset = open("C:/Users/user/Documents/DATA/Documents/Kuliah/SEMESTER 3/Praktikum Big data/saya.txt", "r")

def map_function(text):
    for word in text.split():
        yield (word, 1)

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return result

def mapreduce_wordcount(texts):
    map_result = []
    for text in texts:
        map_result.extend(map_function(text))
    return reduce_function(map_result)


mapped = mapreduce_wordcount(dataset)
for word, count in mapped.items():
      print(f"{word} : {count}")
```

**Custom MapReduce Algorithm**

```python
import csv
from collections import defaultdict

def map_function(data):
    for row in data:
        if len(row) >= 3:  
            try:
                ipa = int(row[1])  #nilai IPA
                ips = int(row[2])  #nilai IPS
                
                # Menampilkan Output
                print(f"Map Output: IPA_total={ipa}, IPS_total={ips}, count=1")
                
                # Yield hasil mapping
                yield ("total_ipa", ipa)
                yield ("total_ips", ips)
                yield ("count", 1)
            except ValueError:
                # Ketikan Syntax ini untuk mengatasi Error
                print(f"Skipping row due to ValueError: {row}")
                continue

def reduce_function(pairs):
    result = defaultdict(int)  
    
    for key, value in pairs:
        result[key] += value
    
    # Menampilkan Hasil
    print("Reduce Output:")
    for key, value in result.items():
        print(f"{key}: {value}")
    
    # Menghitung rata-rata nilai IPA dan nilai IPS
    count = result["count"]
    rata_rata_ipa = result["total_ipa"] / count if count > 0 else 0
    rata_rata_ips = result["total_ips"] / count if count > 0 else 0
    return rata_rata_ipa, rata_rata_ips

# Perintah atau Syntax untuk membaca file nilaisiswa.csv
with open('C:/Users/user/Documents/nilaisiswa.csv', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter=';') 
    next(reader)  
    data = list(reader)

# Menampilkan data dari csv
print("Data nilaisiswa:")
for row in data:
    print(row)

mapped = list(map_function(data))  
rata_rata_ipa, rata_rata_ips = reduce_function(mapped) 

# Menampilkan Hasil Rata-rata nilai IPA dan nilai IPS
print(f"Rata-rata nilai IPA: {rata_rata_ipa}")
print(f"Rata-rata nilai IPS: {rata_rata_ips}")
```
